```python
from google.colab import drive
drive.mount('/content/drive')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import calendar
import matplotlib.gridspec as gridspec
```

```python
df = pd.DataFrame({'Month': [calendar.month_name[i] for i in range(1, 13)],
                   'Sales': np.random.randint(2000, 5000, 12),
                   'Bills': np.random.randint(1000, 4000, 12)},
                  index=range(12))
print(df)
```

```python
fig = plt.figure(figsize=(10, 10))
gs = gridspec.GridSpec(2, 3, height_ratios=[1, 1], width_ratios=[80, 1, 100])


x = np.arange(len(df.Month))
ancho = 0.4
barras = fig.add_subplot(gs[0, :2])
barras.bar(x - ancho/2, df.Sales, width=ancho, label='Ventas', color='blue')
barras.bar(x + ancho/2, df.Bills, width=ancho, label='Gastos', color='red')
plt.ylabel("Monto")
plt.xlabel("Mes")
plt.title("Ventas y Gastos Mensuales")
plt.xticks(x, df.Month)
plt.legend()





linea = fig.add_subplot(gs[1, 0])
linea.plot(df.Month, df.Sales, marker='o', color='red')
linea.plot(df.Month, df.Bills, marker='o', color='blue')
plt.ylabel("Monto")
plt.xlabel("Mes")
plt.title("Ventas y Gastos Mensuales")
plt.legend(['Ventas', 'Gastos'])









dispersion = fig.add_subplot(gs[:, 2])
dispersion.scatter(df.Month, df.Sales, color='green')
dispersion.scatter(df.Month, df.Bills, color='orange')
plt.ylabel("Monto")
plt.xlabel("Mes")
plt.title("Ventas y Gastos Mensuales")
plt.legend(['Ventas', 'Gastos'])

plt.tight_layout()
plt.show()
```
