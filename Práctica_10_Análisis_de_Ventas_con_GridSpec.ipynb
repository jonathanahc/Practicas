```python
from google.colab import drive
drive.mount('/content/drive')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import calendar
import matplotlib.gridspec as gridspec
```

```python
#Genera un DataFrame de ventas con Mes, Ventas, y Gastos.
df = pd.DataFrame({'Month': [calendar.month_abbr[i] for i in range(1, 13)],
                   'Sales': np.random.randint(2000, 5000, 12),
                   'Bills': np.random.randint(1000, 4000, 12)},
                  index=range(12))
print(df)
```

```python
#Usa GridSpec para organizar los subplots de la siguiente manera:
fig = plt.figure(figsize=(10, 10))
gs = gridspec.GridSpec(3, 2, height_ratios=[2,1,2], width_ratios=[2,1])


#Primer gráfico (barra, ocupando dos columnas).
barras = fig.add_subplot(gs[0,:])
barras.bar(df.Month, df.Sales, label="Ventas", color="skyblue")
barras.bar(df.Month, df.Bills, label="Gastos", color="coral", bottom=df.Sales)
plt.ylabel("Monto")
plt.xlabel("Mes")
plt.title("Ventas y Gastos Mensuales")
plt.xticks(rotation=45, ha="right")
plt.legend()


#Segundo gráfico (línea, una columna).
linea = fig.add_subplot(gs[1, 0])
linea.plot(df.Month, df.Sales, marker="o", color="blue", label="Ventas")
linea.plot(df.Month, df.Bills, marker="o", color="red", label="Gastos")
plt.ylabel("Monto")
plt.xlabel("Mes")
plt.title("Ventas y Gastos Mensuales")
plt.xticks(rotation=45, ha="right")
plt.legend()


#Tercer gráfico (dispersión, dos filas).
dispersion = fig.add_subplot(gs[1:, 1])
dispersion.scatter(df.Month, df.Sales, color='green', label="Ventas")
dispersion.scatter(df.Month, df.Bills, color='orange', label="Gastos")
plt.ylabel("Monto")
plt.xlabel("Mes")
plt.title("Ventas y Gastos Mensuales")
plt.xticks(rotation=45, ha='right')
plt.legend()



plt.tight_layout()
plt.show()
```
