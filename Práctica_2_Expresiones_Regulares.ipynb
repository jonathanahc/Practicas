```python
#Monta Google Colab con Drive
from google.colab import drive
drive.mount('/content/drive')

#Importa la librer√≠a Re
import re
```

```python
#Abre el archivo de texto
informe = open('/content/drive/MyDrive/ClassFiles/informe_acciones.txt', encoding='latin-1')
#Crea una lista para la carga de los datos
valores = []
```

```python
#Con el ciclo For, se realiza un recorrido de todo el archivo de texto
for i in informe:
  #Elimina los saltos de l√≠nea y espacios en blanco
  i = i.rstrip()
  #Busca todos los caracteres que est√©n con Valor actual
  if re.search('^Valor actual:.+$', i):
    #Reemplaza la cadena Valor actual por un espacio en blanco
    i = re.sub(r'Valor actual: ', '', i)
    #Inserta los datos en la lista
    valores.append(i)
#Imprime la lista
print(valores)
```
