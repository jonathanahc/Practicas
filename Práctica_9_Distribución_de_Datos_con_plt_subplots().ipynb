```python
from google.colab import drive
drive.mount('/content/drive')
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'Normal': np.random.normal(0, 1, 1000),
                   'Uniforme': np.random.uniform(0, 1, 1000),
                   'Exponencial': np.random.exponential(1, 1000),
                   'Poisson': np.random.poisson(5, 1000)},
                  index=range(1000))
df.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Normal
axes[0, 0].hist(df['Normal'], bins=30)
axes[0, 0].set_title('Normal')

# Uniforme
axes[0, 1].hist(df['Uniforme'], bins=30)
axes[0, 1].set_title('Uniforme')

# Exponencial
axes[1, 0].hist(df['Exponencial'], bins=30)
axes[1, 0].set_title('Exponencial')

# Poisson
axes[1, 1].hist(df['Poisson'], bins=30)
axes[1, 1].set_title('Poisson')

plt.tight_layout()
plt.show()
```
