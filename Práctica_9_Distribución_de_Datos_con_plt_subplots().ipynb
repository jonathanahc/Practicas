```python
from google.colab import drive
drive.mount('/content/drive')
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Genera un DataFrame con Pandas que contenga 1000 valores en cuatro columnas (Normal, Uniforme, Exponencial, Poisson).
df = pd.DataFrame({'Normal': np.random.normal(0, 1, 1000),
                   'Uniforme': np.random.uniform(0, 3, 1000),
                   'Exponencial': np.random.exponential(1, 1000),
                   'Poisson': np.random.poisson(5, 1000)},
                  index=range(1000))
df.head()
```

```python
#Crea una figura con plt.subplots(2, 2) para organizar los histogramas en una cuadrícula de 2x2.
fig, axes = plt.subplots(2, 2, figsize=(12, 8))


# Grafica cada distribución en un subplot usando ax[i, j].hist().
axes[0, 0].hist(df.Normal, bins=30, color="green")
axes[0, 0].set_title("Normal")
axes[0, 0].set_xlabel("Valor")
axes[0, 0].set_ylabel("Frecuencia")


axes[0, 1].hist(df.Uniforme, bins=30, color="orange")
axes[0, 1].set_title("Uniforme")
axes[0, 1].set_xlabel("Valor")
axes[0, 1].set_ylabel("Frecuencia")

axes[1, 0].hist(df.Exponencial, bins=30, color="red")
axes[1, 0].set_title("Exponencial")
axes[1, 0].set_xlabel("Valor")
axes[1, 0].set_ylabel("Frecuencia")


axes[1, 1].hist(df.Poisson, bins=30, color="blue")
axes[1, 1].set_title("Poisson")
axes[1, 1].set_xlabel("Valor")
axes[1, 1].set_ylabel("Frecuencia")


# Agrega títulos a cada subplot y ajusta la disposición con tight_layout().
plt.tight_layout()
plt.show()
```
